```python
 # Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns
%matplotlib inline
```

```python
# Set the file paths for the CSVs using the Path class from the pathlib library
whale_returns_csv_path = Path("whale_returns.csv")
algo_returns_csv_path = Path("algo_returns.csv")
sp500_history_csv_path = Path("SP500_history.csv")
```

```python
#Read data frames and set date as a datetime index
wr_df = pd.read_csv(whale_returns_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
ag_df = pd.read_csv(algo_returns_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
sp_df = pd.read_csv(sp500_history_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
sp_df['Close'] = sp_df['Close'].str.replace('$','')
sp_df.head()
```

```python
sp_df.dtypes
```

```python
sp_df['Close']= sp_df['Close'].astype('float64')
sp_df.dtypes
```

```python
sp_df['Close']= sp_df['Close'].pct_change()
sp_df.head()
```

```python
#sp_daily_returns = sp_df.pct_change()
#sp_daily_returns.head()
```

```python
sp_df.dropna(inplace = True)
```

```python
sp_df.isnull().sum()
```

```python
sp_df.head()
```

```python
sp_df.columns=["S&P 500"]
```

```python
sp_df.head()
```

```python
## Join Whale Returns, Algorithmic Returns, 
#and the S&P 500 Returns into a single DataFrame with columns for 
#each portfolio's returns.
combined_df = pd.concat([wr_df,ag_df,sp_df], axis='columns', join = 'inner')
```

```python
combined_df.head()
```

```python
 # Plot daily returns of all portfolios
   
```

```python
combined_df.plot()
```

```python
combined_df.plot(figsize=(10, 8))
```

```python
 # Calculate cumulative returns of all portfolios
```

```python
cumulative_returns = (1 + combined_df).cumprod() - 1
cumulative_returns.head()
```

```python
# Plot cumulative returns
cumulative_returns.plot(figsize=(10,8))
```

```python
#Determining the risk of each portfolio
```

```python
# Box plot to visually show risk
combined_df.plot.box(figsize = (21,7))
```

```python
# Calculate the daily standard deviations of all portfolios
# Calculate  the daily standard deviation of S&P 500
```

```python
daily_std = combined_df.std()
daily_std
```

```python
sp500_std = sp_df.std()
sp500_std
```

```python
# Determine which portfolios are riskier than the S&P 500
risky_investment = daily_std['S&P 500']<daily_std
risky_investment
```

```python
#calculate the annualized std(252 trading days)
annualized_std = daily_std*np.sqrt(252)
annualized_std
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window

rolling_std_all = combined_df.rolling(window=21).std()
rolling_std_all
```

```python
# Plot the rolling standard deviation
rolling_std_all.plot(figsize=[17,10])
```

```python
# Calculate the correlation
rolling_std_all.corr()
```

```python
# Display the correlation matrix
correlation = rolling_std_all.corr()
sns.heatmap(correlation, vmin=-1,vmax=1)
```

```python
# Calculate covariance of a single portfolio
algo1_covariance = combined_df['Algo 1'].cov(combined_df['S&P 500'])
algo1_covariance




```

```python
# Calculate variance of S&P 500
sp500var=cumulative_returns['S&P 500'].var()
sp500var
```

```python
# Computing beta
algo1_beta = algo1_covariance/sp500var
algo1_beta
```

```python
# Plot beta trend
```

```python
# Use `ewm` to calculate the rolling window
combined_df.ewm(halflife =21).std()
```

```python
# Annualized Sharpe Ratios
sharpe_ratios = (combined_df.mean()*252)/(combined_df.std()*np.sqrt(252))
```

```python
sharpe_ratios
```

```python
sharpe_ratios.plot(kind='bar', title = 'Sharpe Ratios')
```
